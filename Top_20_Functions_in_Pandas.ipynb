14. Which Pandas function is used to write a DataFrame to a CSV file?

The to_csv() function is used to save a DataFrame as a CSV file.

```python
df.to_csv('SalesData.csv', index = False)
```

Why do we use index=False in to_csv()?

By default, Pandas saves the row index (0, 1, 2, ...).


Using index=False prevents the index from being written to the CSV file, making the file cleaner and easier to use in Excel, SQL, or other tools.


15. Slicing with loc and iloc

loc - Uses row/column labels (names)
loc[] selects data using row labels and column names.

iloc - Uses row/column positions (numbers)
iloc[] selects data using row numbers and column numbers.

Select one row

```python
df.loc[7]
```

Select one row and one column

```python
df.loc[2, "Product"]
```

Select multiple rows

```python
df.loc[2:3]
```

Select specific columns

```python
df.loc[:, ["Product", "Quantity"]]
```

```python
df.iloc[7,1]
```

Row position = 7
Column position = 0 (Name)

Select multiple rows



in iloc[], the last row (4) is not included.

```python
df.iloc[1:4]
```

Get only the product abd revenue columns

```python
df.loc[:, ["Product", "Revenue"]]
```

Get rows 5 to 10 and columns 1 to 3

```python
df.iloc[5:11,1:4]
```

Select first two columns

```python
df.iloc[:, 0:2]
```

Which one includes the last value while slicing?


loc[] → Includes the last label.
iloc[] → Excludes the last position.

20. Merging

```python
import pandas as pd

employees = pd.DataFrame({
    "Emp_ID": [101, 102, 103],
    "Name": ["Amit", "Priya", "Rahul"]
})
```

```python
salary = pd.DataFrame({
    "Emp_ID": [101, 102, 104],
    "Salary": [40000, 50000, 60000]
})
```

Merge the dataframes

```python
pd.merge(employees, salary, on="Emp_ID")
```

Types of Merge

There are 4 main types of joins in Pandas.

1. Inner Join (Default)

Returns only matching rows.

```python
pd.merge(employees, salary, on="Emp_ID" , how="inner")
```

2. Left Join

Returns all rows from the left DataFrame and matching rows from the right.

```python
import pandas as pd
pd.merge(employees, salary, on="Emp_ID", how="left")
```

3. Right Join

Returns all rows from the right DataFrame and matching rows from the left.

```python
pd.merge(employees, salary, on="Emp_ID", how="right")
```

4. Outer Join

Returns all rows from both DataFrames.

```python
pd.merge(employees, salary, on="Emp_ID", how="outer")
```
